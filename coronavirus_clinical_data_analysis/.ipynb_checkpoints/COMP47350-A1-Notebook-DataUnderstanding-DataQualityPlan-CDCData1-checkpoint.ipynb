## Check logical integrity of data 
A number of additional checks will be performed to see if the data is reasonable. Depending on the amount of data affected or the analysis that follows, we may decide to drop the rows or replace the values.

### Test 1 Check if any entries have invalid dates (impossible)

```python
# Convert the case_month column to datetime format and check for any invalid dates
# According to DATA.CDC.gov, the data should begin from 2020.01 
# And from the attached file, last updated date was 2022.10.06, check if any case_month was out of the range.
test_1 = []
for year_month in df['case_month']:
    (year, month) = year_month.split("-")
    if 2020 <= int(year) <= 2021:
        if 1 <= int(month) <= 12:
            pass
        else:
            test_1.append(year_month)
    elif int(year) == 2022:
        if 1 <= int(month) <= 10:
            pass
        else:
            test_1.append(year_month)
    else:
        test_1.append(year_month)
        
test_1 = df[df['case_month'].isin(test_1)]

if len(test_1) > 0:
    print(f'there are {len(test_1)} invalid case_month rows')
    print(test_1['case_month'].head())
else:
    print("All case_month formats are correct.")
```

### Test 2 Check that all state abbreviations are correct

```python
# List all state abbreviations and store as valid_state_abbr
valid_state_abbr = ['AL', 'AK', 'AZ', 'AR', 'AS', 'CA', 'CO', 'CT', 'DE','DC', 'FL',
              'GA', 'GU', 'HI', 'ID', 'IL', 'IN', 'IA', 'KS', 'KY', 'LA', 'ME',
              'MD', 'MA', 'MI', 'MN', 'MS', 'MO', 'MT', 'NE', 'NV', 'NH', 'NJ',
              'NM', 'NY', 'NC', 'ND', 'MP', 'OH', 'OK', 'OR', 'PA', 'PR', 'RI', 
              'SC', 'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'VI', 'WA', 'WV', 'WI', 'WY']

# Check if all the state abbreviations are correct
invalid_states = df[~df['res_state'].isin(valid_state_abbr)]['res_state'].unique()
if len(invalid_states) == 1 and np.isnan(invalid_states[0]):
    print("All the state abbreviations are correct.")
elif len(invalid_states) > 0:
    print(f"Found {len(invalid_states)} abbreviations for invalid states: {invalid_states}")
else:
    print("All the state abbreviations are correct.")
```

## 2. Data Understanding: Data Quality Plan for the cleaned CSV file.

In this section we list all features and the observations and actions involving those features, as a step towards obtaining a cleaner dataset.

### The initial list of issues as identified in the **Data_Quality_Report_Initial_Findings.pdf**:

- Logical tests in part 1 had some failures which should be handled.
- Special value â€˜missingâ€™ and 'unknown' appeared in many rows covering multiple features. Given that values of Unknown and Missing is also invalid as NAN value, convert these values to NAN value for further analysis.
- Special negative values appears in 2 features. *case_positive_specimen_interval* & *case_onset_interval*. It indicates positive specific date/onset date is earlier than cdc case earliest date, which makes no sense.
- Different scales used for similar features - Categorical features *res_county* & *county_fips_code* are measuring the same event and are using the same numerical mappings. One mapping needs to be picked for consistency.
- Presence of outliers - There are a significant number of outliers present across a range of different features. They initially look implausible and will need to be investigated further.

#### **New feature 1** 

- The new feature "res_region" is a categorical column that groups the states into five regions of the United States. 
- This feature helps to understand the geographic distribution of the data and to explore any regional differences in the variables of interest.

```python
# Python dictionary mapping two letter state abbreviations to their respective regions of the country (i.e. Midwest, North East, etc.) 
# N - North East W - West M - Mid West S - South O - Other

states_to_region = {
        'AK': 'O',
        'AL': 'S',
        'AR': 'S',
        'AS': 'O',
        'AZ': 'W',
        'CA': 'W',
        'CO': 'W',
        'CT': 'N',
        'DC': 'N',
        'DE': 'N',
        'FL': 'S',
        'GA': 'S',
        'GU': 'O',
        'HI': 'O',
        'IA': 'M',
        'ID': 'W',
        'IL': 'M',
        'IN': 'M',
        'KS': 'M',
        'KY': 'S',
        'LA': 'S',
        'MA': 'N',
        'MD': 'N',
        'ME': 'N',
        'MI': 'W',
        'MN': 'M',
        'MO': 'M',
        'MP': 'O',
        'MS': 'S',
        'MT': 'W',
        'NA': 'O',
        'NC': 'S',
        'ND': 'M',
        'NE': 'W',
        'NH': 'N',
        'NJ': 'N',
        'NM': 'W',
        'NV': 'W',
        'NY': 'N',
        'OH': 'M',
        'OK': 'S',
        'OR': 'W',
        'PA': 'N',
        'PR': 'O',
        'RI': 'N',
        'SC': 'S',
        'SD': 'M',
        'TN': 'S',
        'TX': 'S',
        'UT': 'W',
        'VA': 'S',
        'VI': 'O',
        'VT': 'N',
        'WA': 'W',
        'WI': 'M',
        'WV': 'S',
        'WY': 'W'
}


df["res_region"] = df["res_state"].replace(states_to_region)
df['res_region'] = df['res_region'].astype('category')
value_counts = df["res_region"].value_counts()

fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(value_counts.index, value_counts.values)
ax.set_xlabel("Region")
ax.set_ylabel("Count")

# add count values as annotations to each bar
for i, v in enumerate(value_counts.values):
    ax.text(i, v+1, str(v), ha='center', fontsize=10)

plt.show()
```
